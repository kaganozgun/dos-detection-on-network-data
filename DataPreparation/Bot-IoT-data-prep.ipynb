```python
import pandas as pd
pd.set_option('display.expand_frame_repr', False)
from sklearn import preprocessing
import glob
import os
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
index = 0
prev = 1
for a in [6,11,16,21,26,31,36,41,46,51,56,61,66,71,75]:
  colnames = ["pkSeqID","stime","flgs","proto","saddr","sport","daddr","dport","pkts","bytes","state","ltime","seq","dur","mean","stddev","smac","dmac","sum","min","max","soui","doui","sco","dco","spkts","dpkts","sbytes","dbytes","rate","srate","drate","attack","category","subcategory"]

  base_path = '/content/drive/MyDrive/Okul/İTÜ/Bilgisayar Mühendisliği Yüksek Lisans/Yüksek Lisans/YL Tez/Development/Bot-Iot/CSV/Entire Dataset/'
  df_list = []
  for i in range(prev,a):
    prev = a
    print(i)
    df = pd.read_csv(base_path + "UNSW_2018_IoT_Botnet_Dataset_"+ str(i) +".csv", names=colnames, header=None)
    df_list.append(df)
  data = pd.concat(df_list, ignore_index=True)
  print(data.shape)
  df = data.drop(['stime', 'saddr', 'sport', 'daddr','dport','subcategory','smac','soui','doui','sco','dmac','dco','pkSeqID','flgs'], axis=1)
  del data
  df_dos = df.loc[df['category'] == 'DoS']
  print(df_dos.shape)
  df_normal = df.loc[df['category'] == 'Normal']
  print(df_normal.shape)
  attcak_row_size = df_normal.shape[0]*0.3/0.7
  if(df_dos.shape[0] > 0):
    sampled_dos = df_dos.sample(int(attcak_row_size),replace=True)
    sampled_df = pd.concat([sampled_dos, df_normal])
    del sampled_dos
    del df_normal
    sampled_df = sampled_df.sort_values(by=['ltime'], ascending=True)
    print(sampled_df.shape)
    sampled_df_label = sampled_df['attack']
    sampled_df = sampled_df.drop(['attack', 'category','ltime'], axis=1)
    proto_categorical_cols = pd.get_dummies(sampled_df['proto'], prefix='proto', prefix_sep='_', dummy_na=False, columns=['proto'], sparse=False, drop_first=False, dtype=None)
    state_categorical_cols = pd.get_dummies(sampled_df['state'], prefix='state', prefix_sep='_', dummy_na=False, columns=['state'], sparse=False, drop_first=False, dtype=None)
    sampled_df = sampled_df.drop(['state', 'proto'], axis=1)
    sampled_df_2 = pd.concat([sampled_df, state_categorical_cols, proto_categorical_cols],axis=1)
    x = sampled_df_2.values #returns a numpy array
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    df = pd.DataFrame(x_scaled)
    df2 = df.fillna(0)
    df2.reset_index(inplace=True, drop=True)
    sampled_df_label.reset_index(inplace=True, drop=True)
    final_df = pd.concat([df2, sampled_df_label], axis=1)
    print(final_df.shape)
    final_base_path = '/content/drive/MyDrive/Okul/İTÜ/Bilgisayar Mühendisliği Yüksek Lisans/Yüksek Lisans/YL Tez/Development/Bot-Iot/Sampled/'
    final_df.to_csv (final_base_path + '30-sampled-data_' + str(index) + '.csv', header=True)
    index = index + 1
```
