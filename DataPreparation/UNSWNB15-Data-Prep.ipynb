```python
import pandas as pd
pd.set_option('display.expand_frame_repr', False)
from sklearn import preprocessing
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
colnames = ['srcip','sport','dstip','dsport','proto','state','dur','sbytes','dbytes','sttl','dttl','sloss','dloss'
                ,'service','Sload','Dload','Spkts','Dpkts','swin','dwin','stcpb','dtcpb','smeansz','dmeansz','trans_depth',
                'res_bdy_len','Sjit','Djit','Stime','Ltime','Sintpkt','Dintpkt','tcprtt','synack','ackdat',
                'is_sm_ips_ports','ct_state_ttl','ct_flw_http_mthd','is_ftp_login','ct_ftp_cmd','ct_srv_src',
                'ct_srv_dst','ct_dst_ltm','ct_src_ ltm','ct_src_dport_ltm','ct_dst_sport_ltm','ct_dst_src_ltm',
                'attack_cat','Label']
df_1 = pd.read_csv("/content/drive/MyDrive/Okul/İTÜ/Bilgisayar Mühendisliği Yüksek Lisans/Yüksek Lisans/YL Tez/Development/UNSW-NB15/dataset/UNSW-NB15_1.csv", names=colnames, header=None)
df_2 = pd.read_csv("/content/drive/MyDrive/Okul/İTÜ/Bilgisayar Mühendisliği Yüksek Lisans/Yüksek Lisans/YL Tez/Development/UNSW-NB15/dataset/UNSW-NB15_2.csv", names=colnames, header=None)
df_3 = pd.read_csv("/content/drive/MyDrive/Okul/İTÜ/Bilgisayar Mühendisliği Yüksek Lisans/Yüksek Lisans/YL Tez/Development/UNSW-NB15/dataset/UNSW-NB15_3.csv", names=colnames, header=None)
df_4 = pd.read_csv("/content/drive/MyDrive/Okul/İTÜ/Bilgisayar Mühendisliği Yüksek Lisans/Yüksek Lisans/YL Tez/Development/UNSW-NB15/dataset/UNSW-NB15_4.csv", names=colnames, header=None)
data = pd.concat([df_1, df_2, df_3, df_4], ignore_index=True)
```

```python
data.shape
```

```python
dos = data.loc[data['attack_cat'] == 'DoS']
dos.shape
```

```python
normal = data.loc[data['Label'] == 0]
normal.shape
```

```python
df = data.drop(['srcip', 'sport', 'dstip', 'dsport'], axis=1)
df
```

```python
is_NaN = df[['attack_cat']].isnull()
row_has_NaN = is_NaN.any(axis=1)
normal_df = df[row_has_NaN]
normal_df
```

```python
df_ddos = df.loc[data['attack_cat'] == 'DoS']
df_ddos
```

```python
normal_attack_rate = 0.3
attcak_row_size = df_ddos.shape[0]*0.7/0.3
attcak_row_size
```

```python
sampled_normal = normal_df.sample(int(attcak_row_size),replace=True)
sampled_normal.shape
```

```python
sampled_df = pd.concat([sampled_normal, df_ddos])
sampled_df = sampled_df.sort_index()
sampled_df_label = sampled_df['Label']
sampled_df = sampled_df.drop(['attack_cat', 'ct_ftp_cmd','Label'], axis=1)
sampled_df
```

```python
service_categorical_cols = pd.get_dummies(sampled_df['service'], prefix='service', prefix_sep='_', dummy_na=False, columns=['service'], sparse=False, drop_first=False, dtype=None)
state_categorical_cols = pd.get_dummies(sampled_df['state'], prefix='state', prefix_sep='_', dummy_na=False, columns=['state'], sparse=False, drop_first=False, dtype=None)
proto_categorical_cols = pd.get_dummies(sampled_df['proto'], prefix='proto', prefix_sep='_', dummy_na=False, columns=['proto'], sparse=False, drop_first=False, dtype=None)
```

```python
sampled_df = sampled_df.drop(['service', 'state', 'proto'], axis=1)
sampled_df_2 = pd.concat([sampled_df, service_categorical_cols, state_categorical_cols, proto_categorical_cols],axis=1)
sampled_df_2.shape
```

```python
x = sampled_df_2.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
df2 = df.fillna(0)
```

```python
df2.reset_index(inplace=True, drop=True)
sampled_df_label.reset_index(inplace=True, drop=True)
final_df = pd.concat([df2, sampled_df_label], axis=1)
final_df
```

```python
final_df.to_csv (r'/content/drive/MyDrive/Okul/İTÜ/Bilgisayar Mühendisliği Yüksek Lisans/Yüksek Lisans/YL Tez/Development/UNSW-NB15/dataset/30-sampled-data.csv', header=True)
```

```python
base_path = "/content/drive/MyDrive/Okul/İTÜ/Bilgisayar Mühendisliği Yüksek Lisans/Yüksek Lisans/YL Tez/Development/UNSW-NB15/dataset"
data = pd.read_csv(f"{base_path}/30-sampled-data.csv", index_col=0)
data = data.fillna(0)
X = data.iloc[:, 1:-1]
y = data.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=69)
```
