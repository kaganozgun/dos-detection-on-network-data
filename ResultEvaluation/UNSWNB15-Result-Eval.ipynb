```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
base_path = "/content/drive/MyDrive/Okul/İTÜ/Bilgisayar Mühendisliği Yüksek Lisans/Yüksek Lisans/YL Tez/Development/UNSW-NB15"
rnn_time_series = np.genfromtxt(f"{base_path}/model/rnn-time-series-y-pred.csv", delimiter=',')
lstm = np.genfromtxt(f"{base_path}/model/lstm-y-pred.csv", delimiter=',')
rnn = np.genfromtxt(f"{base_path}/model/rnn-y-pred.csv", delimiter=',')
y_test = np.genfromtxt(f"{base_path}/dataset/30-sampled-y-test.csv", delimiter=',')
```

```python
rnn_time_series_df = pd.DataFrame(rnn_time_series, columns = ['rnn_time_series'])
lstm_df = pd.DataFrame(lstm, columns = ['lstm'])
rnn_df = pd.DataFrame(rnn, columns = ['rnn'])
result_df = pd.concat([rnn_time_series_df, lstm_df, rnn_df], axis=1)
result_df
```

```python
col_list= ['rnn_time_series', 'lstm', 'rnn']
result_df['Sum'] = result_df[col_list].sum(axis=1)
result_df
```

```python
result_df['final'] = result_df['Sum'].apply(lambda x: 1 if x == 3 else 0)
result_df
```

```python
final = result_df['final'].to_numpy()
cm = confusion_matrix(y_test, final)
print(cm)
print(classification_report(y_test, final))
print(f"Accuracy: {accuracy_score(y_test, final)}")
print(f"False Positive rate: {1 - (cm[0][1] / len(y_test))}")
print(f"False Negative rate: {1 - (cm[1][0] / len(y_test))}")
```
